# Substitution Ciphers

## Introduction

The Caesar cipher and substitution cipher are classic encryption methods used to disguise messages. The Caesar cipher shifts each letter by a fixed number of places in the alphabet, making it easy to implement but also easy to break through brute force.
The substitution cipher is more flexible, replacing each letter with a different one based on a key. While stronger than the Caesar cipher, it can still be cracked using frequency analysis, which examines letter usage patterns.
Both ciphers played a key role in early cryptography but are now considered weak due to modern decryption techniques.

## Caesar Cipher

The Caesar cipher is a simple encryption method that shifts each letter in the plaintext by a fixed number of positions in the alphabet. For example, with a shift of 3, "A" becomes "D," "B" becomes "E," and so on. It is easy to implement but also easy to break using brute force since there are only 26 possible shifts. Despite its simplicity, it was historically used for secure communication and remains a fundamental concept in cryptography.

### Encryption

```python
def caesar_encrypt(plaintext, shift=0):
    
    al = "abcdefghijklmnopqrstuvwxyz"
    ciphertext = ""

    for i  in range(len(plaintext)):
        
        if plaintext[i].islower(): 
            
            plain_index = ord(plaintext[i]) - ord("a")               
            ciphertext += al[(plain_index + shift) % 26]
            
        else:
            ciphertext = ciphertext + plaintext[i]
            
    return ciphertext
```

```python
# code snippet to test the implementation of the encryption function
plaintext = 'hello!' 
ciphertext = caesar_encrypt(plaintext, shift=4)

print(plaintext, '->', ciphertext) # expected output 'hello! -> lipps!'
```

### Decryption

* Caesar cipher decryption works by shifting each letter back by the same fixed number used for encryption. For example, if a shift of +3 was applied during encryption, decryption shifts the letters -3 to restore the original text.

```python
def caesar_decrypt(ciphertext, shift=0):
    letters = "abcdefghijklmnopqrstuvwxyz"
    plaintext = ""

    for i  in range(len(ciphertext)):
        
        if ciphertext[i].islower(): 
            
            plaintext += letters[(letters.find(ciphertext[i]) - shift) % 26]
    
        else:            
            plaintext += ciphertext[i]
            
    return plaintext
```

```python
# code snippet to test the implementation of the decryption function
ciphertext = 'lipps!' # 'hello!' encoded with shift=4
plaintext = caesar_decrypt(ciphertext, shift=4)

print(ciphertext, '->', plaintext)  # expected output 'lipps! -> hello!'
```

## Simple Substitution Cipher

* A simple substitution cipher replaces each letter in the plaintext with a different letter from the alphabet based on a predefined key. Unlike the Caesar cipher, which shifts letters by a fixed amount, this method allows any arbitrary letter-to-letter mapping, making it more secure. However, it is still vulnerable to frequency analysis, as common letters in the plaintext will appear frequently in the ciphertext.

### Encryption

* A substitution cipher encrypts a plaintext message by replacing each lowercase letter with a corresponding character from a given mapping. The function verifies whether a character is lowercase and substitutes it accordingly. Non-alphabetic characters (such as spaces, punctuation, and uppercase letters) remain unchanged, preserving the original structure of the message.

```python
def substitution_encrypt(plaintext, mapping):
    ciphertext = ""

    for char in plaintext:
        if char in mapping:  
            ciphertext += mapping[char]
        else:
            ciphertext += char 

    return ciphertext
```

* Here, I utilized the built-in zip function to pair the standard alphabet with our custom substitution cipher, creating a dictionary that maps each letter to its encrypted counterpart. This allows for efficient letter substitution during the encryption process.

```python
# code snippet to test the implementation of the encryption function
plaintext = 'hello!'

letters =     "abcdefghijklmnopqrstuvwxyz"
letters_enc = "sgrepyuaotiwdfqhjklxzcvbmn"
#mapping = {'h': 'a', 'e': 'p', 'l': 'w', 'o': 'q'}

mapping = {key : value for key , value in zip(letters,letters_enc)}
ciphertext = substitution_encrypt(plaintext, mapping)


print(plaintext, '->', ciphertext) # expected output 'hello! -> apwwq!'
```

### Decryption

* The decryption process for a substitution cipher works by reversing the letter substitutions using a remapping. It takes the ciphertext and the original encryption mapping, then creates a remapping dictionary where the keys and values are swapped. Each letter in the ciphertext is then replaced with its corresponding letter from the remapping, effectively restoring the original plaintext. Non-alphabetic characters remain unchanged, preserving the structure of the message.

```python
def substitution_decrypt(ciphertext, mapping):
    plaintext = ""
    # Reverse mapping
    remapping = {v: k for k, v in mapping.items()}  

    for char in ciphertext:
        if char in remapping:  
            plaintext += remapping[char]
        else:
            plaintext += char  
            
    return plaintext
```

### Frequency Analysis Attack

* A frequency analysis attack exploits the fact that certain letters appear more frequently in a language than others. By analyzing the frequency of letters in the ciphertext and comparing it to standard letter frequencies in English, we can make educated guesses about letter substitutions and reconstruct the plaintext.
* The attack is more effective on longer messages because letter frequency patterns become clearer with more data.

#### English Letters Distribution

```python
# function to infer letter frequency distribution from a text
def letter_distribution(text):
    alphabet = "abcdefghijklmnopqrstuvwxyz"
    letter_counts = {char: 0 for char in alphabet}  
    total_letters = 0

    for character in text:
        if character in letter_counts: 
            letter_counts[character] += 1
            total_letters += 1

    # Compute the frequency distribution
    frequency_distribution = (
        {char: count / total_letters for char, count in letter_counts.items() if count > 0}
        if total_letters > 0 else {}
    )

    return frequency_distribution
```

## Affine Cipher

### Introduction

In this task, we will work with the Affine cipher, a simple encryption method that uses two keys to encrypt and decrypt messages. It is defined by the equation y = (a*x + b) mod m, where a and b are keys, x is the numerical value of the plaintext letter, and m is the size of the alphabet. For decryption, the modular inverse of a is used to reverse the transformation.

We will:

* Encrypt a message using the affine cipher.
* Decrypt the ciphertext back into plaintext using the appropriate keys.
* Load a ciphertext from a file and attempt to break the cipher by testing different key combinations.


### Encryption

* In the encryption part of the code, we implement the Affine cipher’s encryption algorithm. For each letter in the plaintext, we convert it to a numerical value and apply the formula   y=(a×x+b) modm ,
* * where: 𝑎 and 𝑏 are the encryption keys, 𝑥 is the numeric equivalent of the plaintext letter, 𝑚 is the size of the alphabet (26 for English letters).
    
After applying this formula, we convert the resulting value back to a letter, forming the ciphertext. Non-alphabetic characters (such as spaces and punctuation) remain unchanged.
The encryption process in the code is implemented in a loop that handles each character, ensuring proper transformation of the plaintext into ciphertext.

```python
from math import gcd
def affine_encrypt(plaintext, a, b):
    al = "abcdefghijklmnopqrstuvwxyz"
    ciphertext = ""

    for char in plaintext:
        if char.islower():  
            x = al.find(char)
            ciphertext += al[(a * x + b) % 26]
        else:
            ciphertext += char  

    return ciphertext
```

```python
plaintext = 'hello world!'
a, b = 3, 1

ciphertext = affine_encrypt(plaintext, a, b)
print(plaintext, '->', ciphertext) # expected output 'hello world! -> wniir praik!'
```

### **Decryption**

In the decryption part of the code, we reverse the Affine cipher’s encryption process. The key step in decryption is finding the modular inverse of 
a, which is used to undo the effect of the encryption formula. For each letter in the ciphertext, we use this inverse along with the additive key 
b to transform the ciphertext back into plaintext.

The modular inverse of a is calculated first, and once found, it is applied to each character in the ciphertext. The characters are then converted back into their corresponding plaintext letters, while non-alphabetic characters remain unchanged.

The decryption process is implemented in a loop that processes each character individually, ensuring that the ciphertext is correctly reversed to recover the original plaintext.

```python
def affine_decrypt(ciphertext, a, b):
    al = "abcdefghijklmnopqrstuvwxyz"
    
    # Find the modular inverse of 'a' modulo 26
    a_bar = None
    for i in range(26):
        if (a * i) % 26 == 1:
            a_bar = i
            break
    
    if a_bar is None:
        raise ValueError(f"No modular inverse found for a = {a} in modulus 26.")
    
    plaintext = ""
    
    for char in ciphertext:
        if char.islower():  
            x = al.find(char)
            plaintext += al[(a_bar * (x - b)) % 26]
        else:
            plaintext += char  
    
    return plaintext
```

### Breaking Cipher

```python
def affine_break(ciphertext):
    a_inverse = [1, 9, 15, 3, 19, 7, 23, 11, 5, 17, 25]
    
    refrence_words = [" the ", " in ", " a ", " an ", " on ", " with"]  
    al = "abcdefghijklmnopqrstuvwxyz"
    
    for b in range(26):  # Try all values for 'b' (shift)
        for a_bar in a_inverse:  # Try all inverses of 'a'
            plaintext = ""
            
            # Decrypt the ciphertext with the current 'a_bar' and 'b'
            for char in ciphertext:
                if char.islower():  
                    ciphertext_index = al.find(char)
                    plaintext += al[(a_bar * (ciphertext_index - b)) % 26]
                else:
                    plaintext += char  

            if all(word in plaintext for word in refrence_words):
                return plaintext, b, a_bar  

    return None  

result = affine_break(ciphertext)

if result:
    decrypted_text, b, a_bar = result
    print("Shift key (b):", b)
    print("Multiplicative key (a_bar):", a_bar)
    print("Decrypted text:", decrypted_text)
```
